```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/excess_mortality.csv"
```

```python
df =pd.read_csv(path)
```

```python
total_deaths = df.groupby('location')[['deaths_2020_all_ages', 'deaths_2019_all_ages']].sum()
```

```python
mortality_rate_2020 = total_deaths['deaths_2020_all_ages'] / total_deaths['deaths_2019_all_ages']
mortality_rate_2019 = total_deaths['deaths_2019_all_ages'] / total_deaths['deaths_2019_all_ages']
```

```python
lowest_mortality = []
for i in range(len(total_deaths)):
    if mortality_rate_2020[i] < mortality_rate_2019[i]:
        lowest_mortality.append('2020')
    else:
        lowest_mortality.append('2019')
```

```python
total_deaths['lowest_mortality'] = lowest_mortality
```
