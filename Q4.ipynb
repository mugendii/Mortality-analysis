QUESTION 4
For each country, find which year had the lowest mortality

```python
import pandas as pd
from statistics import mean
```

```python
data = 'excess_mortality.csv'
df= pd.read_csv(data)
df
```

```python
Austria_df =df.loc[df['location'] == 'Austria']
Austria_df
```

```python
#removing unwanted columns
Austria_df = Austria_df.drop(['p_scores_all_ages','p_scores_15_64','p_scores_65_74','p_scores_75_84','p_scores_85plus','average_deaths_2015_2019_all_ages','Week'], axis=1)
Austria_df
```

```python
Austria_df.mean()
```

```python
Average_deaths = Austria_df.mean()
Lowest_mortality = Average_deaths.min()
print(Lowest_mortality)
min_col = Average_deaths.idxmin()

# Get the actual minimum value
min_val = df[min_col].min()

print(f"The column with the lowest mean is '{min_col}', with a mean of {Average_deaths[min_col]} and a minimum value of {min_val}.")
```
